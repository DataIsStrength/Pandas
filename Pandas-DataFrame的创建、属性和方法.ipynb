Pandas-DataFrame的创建、属性和方法

```python
#导入NumPy包
import numpy as np
#导入Pandas包
import pandas as pd

'''
Pandas是Panel data analysis（面板数据分析）的缩写。
Pandas结合了NumPy和Matplotlib的功能，并且具有自己独特的数据结构。
'''
```

```python
'''
Pandas的数据结构有Series，DataFrame和Panel（已移除）,其中最常用的是DataFrame。
DataFrame是既有行索引，又有列索引的二维数组。
'''

#用NumPy数组创建DataFrame
stock_change=np.random.normal(0,1,(10,5))
df1=pd.DataFrame(stock_change)
print(df1)
```

```python
#添加行索引
stock=['股票{}'.format(i) for i in range(10)]
df2=pd.DataFrame(stock_change,index=stock)
print(df2)

'''
注意：如果用一个DataFrame创建另一个DataFrame，那么创建的DataFrame所有元素为缺失值。
'''
```

```python
#pd.date_range()
date=pd.date_range(start='20180101', periods=5, freq='B')
'''
pd.date_range()的参数：
（1）start：起始时间；
（2）end：结束时间；
（3）periods：期数；
（注意：start、end和periods这3个参数至少应该设置2个）
（4）freq：频率，常见取值如下：
①交易日频率：B；
②日历日频率：D；
③周频率：W；
④季度频率：Q；
⑤年频率：A,Y。
'''

#添加列索引
df3=pd.DataFrame(stock_change,index=stock,columns=date)
print(df3)
```

```python
#DataFrame的属性

#形状
print(df3.shape)
#行索引
print(df3.index)
#列索引
print(df3.columns)
#值（ndarray）
print(df3.values)
```
