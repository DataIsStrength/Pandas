Pandas-DataFrame的合并

```python
#导入NumPy包
import numpy as np
#导入Pandas包
import pandas as pd
```

```python
#导入数据
stock=pd.read_csv('stock_day.csv')
#将p_change列独热编码
p_change=stock['p_change']
bins=[-100,-7,-5,-3,0,3,5,7,100]
stock_change=pd.cut(p_change,bins)
stock_change=pd.get_dummies(stock_change, prefix='rise')
```

```python
#按方向合并（pd.concat()）

'''
pd.concat的轴参数：默认axis=0，即垂直合并，当axis=1时水平合并。
'''

#水平合并
pd.concat([stock,stock_change],axis=1)
```

```python
#按索引合并（pd.merge()）

'''
pd.merge()的参数：
（1）how：连接方式：
①inner：内连接，合并后只保留左右表索引列数据相同的记录（默认值，最常用）；
②left：左连接，按照左表的索引合并，只保留右表中可以并入左表索引的记录；
③right：右接连，按照右表的索引合并，只保留左表中可以并入右表索引的记录；
④outer：外接连，合并后保留所有记录。
（2）on：索引列。
'''

#导入左右表数据
left=pd.DataFrame({'key1':['K0','K0','K1','K2'],
                   'key2':['K0','K1','K0','K1'],
                      'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['K0','K1','K1','K2'],
                    'key2':['K0','K0','K0','K0'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
#以key1和key2为索引列内连接
pd.merge(left,right,how='inner',on=['key1','key2'])
```

```python
#以key1和key2为索引列左连接
pd.merge(left,right,how='left',on=['key1','key2'])
```

```python
#以key1和key2为索引列右连接
pd.merge(left,right,how='right',on=['key1','key2'])
```

```python
#以key1和key2为索引列外连接
pd.merge(left,right,how='outer',on=['key1','key2'])
```
