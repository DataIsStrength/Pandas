Pandas-DataFrame索引设置

```python
#导入NumPy包
import numpy as np
#导入Pandas包
import pandas as pd
```

```python
#用NumPy数组创建DataFrame
#生成随机数据
stock_change=np.random.normal(0,1,(10,5))
#设置行索引
stock=['股票{}'.format(i) for i in range(10)]
#设置列索引
date=pd.date_range(start='20180101',periods=5,freq='B')
#创建DataFrame
df=pd.DataFrame(stock_change,index=stock,columns=date)
print(df)
```

```python
#修改列索引
stock_=['股票_{}'.format(i) for i in range(10)]
df.index=stock_
print(df)

'''
DataFrame的索引只可以整体修改，不可以单独修改其中的一项。
'''
```

```python
#重置索引并保留原索引列
df.reset_index(drop=False)
```

```python
#重置索引不保留原索引列
df.reset_index(drop=True)
```

```python
df=pd.DataFrame({'month':[1,4,7,10],
                  'year':[2012,2014,2013,2014],
                  'sale':[55,40,84,31]})
print(df)
```

```python
#设置新索引

#设置新的索引并删除原索引列
df1=df.set_index('month',drop=True)
print(df1)
#设置多个索引
df2=df.set_index(['year','month'],drop=True)
print(df2)
print(df2.index)
print(df2.index.names)
print(df2.index.levels)
```
