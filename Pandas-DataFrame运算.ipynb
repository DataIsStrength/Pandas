Pandas-DataFrame运算

```python
#导入NumPy包
import numpy as np
#导入Pandas包
import pandas as pd
```

```python
data=pd.read_csv('stock_day.csv')
#删除多余字段
data=data.drop(['ma5','ma10','ma20','v_ma5','v_ma10','v_ma20'],axis=1)
```

```python
data
```

```python
#算术运算

#加
print(data+1)
#或者
print(data.add(1))

#减
print(data-1)
#或者
print(data.sub(1))

#乘
print(data*10)

#除
print(data/10)

#两个Series之间的运算
print(data['close'].sub(data['open']))

'''
Series的算术运算和DataFrame类似。
'''
```

```python
#逻辑运算

#筛选p_change>2的记录
print(data[data['p_change']>2])

#筛选p_change>2并且low>15的记录
print(data[(data['p_change']>2)&(data['low']>15)])
#或者
print(data.query('p_change>2&low>15'))

#筛选turnover=2.39或4.19的数据
print(data[data['turnover'].isin([2.39,4.19])])
```

```python
#统计运算

#统计描述
print(data.describe())

#列最大值
print(data.max(axis=0))
'''
常用的统计运算包括：max、min、mean、median、var、std。 
'''

#列最大值的索引
print(data.idxmax(axis=0))

#Series中最大值的索引序号
print(np.argmax(data['close']))
```

```python
#累计统计函数

'''
累计统计函数包括：
①cumsum：计算前1/2/3.../n个数的和
②cummax：计算前1/2/3.../n个数的最大值
③cummin：计算前1/2/3.../n个数的最小值
④cumprod：计算前1/2/3.../n个数的积
'''

#将p_change按时间排序后累加并绘图
data['p_change'].sort_index().cumsum().plot()
```

```python
#自定义运算

#.apply()的第一个参数是函数（可以用匿名函数表示），第二个参数是轴参数
#每一列最大值和最小值的差
print(data.apply(lambda x:x.max()-x.min(),axis=0))
```
