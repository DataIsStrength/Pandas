Pandas-基本数据操作

```python
#导入NumPy包
import numpy as np
#导入Pandas包
import pandas as pd
```

```python
data=pd.read_csv('stock_day.csv')
```

```python
data
```

```python
#删除多余字段
data=data.drop(['ma5','ma10','ma20','v_ma5','v_ma10','v_ma20'],axis=1)
print(data)
```

```python
#索引

#索引第2行第1列
#提取Series然后对Series进行索引
print(data['open']['2018-02-26'])
#.loc()索引
print(data.loc['2018-02-26','open'])
#.iloc()索引
print(data.iloc[1,0])
```

```python
#序号和字段名混合索引
['open', 'close', 'high', 'low']
#方法1
print(data.loc[data.index[0:4],['open', 'close', 'high', 'low']])
#方法2
print(data.iloc[0:4,data.columns.get_indexer(['open', 'close', 'high', 'low'])])
```

```python
#赋值

#整列赋值
data['open']=100
#单个值赋值
data.iloc[1,0]=222
print(data)
```

```python
#DataFrame排序

#按字段排序
#按high和p_change字段降序排列
data1=data.sort_values(by=['high','p_change'],ascending=False)
print(data1)

#按索引列排序
data2=data.sort_index()
print(data2)
```

```python
#Series排序

sr=data['p_change']
#升序排序
sr=sr.sort_values(ascending=True)
print(sr)
```
