Pandas-读取和存储文件

```python
#导入NumPy包
import numpy as np
#导入Pandas包
import pandas as pd
```

```python
#读取csv文件

#读取csv文件并选择字段
data1=pd.read_csv('stock_day.csv',usecols=['high','low','open','close'])
print(data1)
#读取没有字段名的csv文件并设置字段名
data2=pd.read_csv('stock_day2.csv',names=['open','high','close','low','volume','price_change','p_change','ma5','ma10','ma20','v_ma5','v_ma10','v_ma20','turnover'])
print(data2)
```

```python
#保存数据到csv文件

'''
.to_csv()的参数：
①columns：需要保存的字段名；
②index：是否保留索引列，默认值为True；
③header：是否保留字段名，默认值为True；
④mode：取值为a时为追加模式，在原数据表格下方续写。
'''

#保存open列前10行，并且不保留索引列
data1[:10].to_csv('open.csv',columns=['open'],index=False,header=False)
```

```python
#读取hdf5文件

'''
hdf5文件以键值对的形式存储三维数据，每个键对应一个二维数据表格，对于含有多个键值对的hdf5文件，在读取时需要指定一个特定的键。
'''

#（对于单个键值对的hdf5文件，可以不设置key参数）
day_open=pd.read_hdf('day_open.h5')
day_close=pd.read_hdf('day_close.h5')
print(day_open)
print(day_close)
```

```python
#存储hdf5文件

#将不同的DataFrame存储到hdf5文件的不同键值对
day_open.to_hdf('day.h5',key='open')
day_close.to_hdf('day.h5',key='close')
```

```python
#读取json文件

'''
pd.read_json()的参数：
①orient：格式，一般设置为'records'；
②lines：是否以一行为一条记录，默认值为False，一般设置为True。
'''

json_data=pd.read_json('Sarcasm_Headlines_Dataset.json',orient='records',lines=True)
print(json_data)
```

```python
#存储json文件

#（.to_json()的参数设置和pd.read_json()类似）
json_data.to_json('json_data.json',orient='records',lines=True)
```
